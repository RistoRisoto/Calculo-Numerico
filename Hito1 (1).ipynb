```python
import matplotlib.pyplot as plt
import math as m
import numpy as np
def draw(f,n=50,li=-1,lf=1):
    xl=np.linspace(li,lf,n)
    yl=[]
    for i in range(len(xl)):
        yl.append(f(xl[i]))
    plt.plot(xl,yl)

def coords(f,n=50,li=-1,lf=1):
    xl=np.linspace(li,lf,n)
    yl=[]
    for i in range(len(xl)):
        yl.append(f(xl[i]))
    return xl,yl
    
def plotI(f,x0):
    y0=[]
    for i in range(len(x0)):
        y0.append(I(f,x0[i]))
    plt.plot(x0,y0,'o')
    def If(x):
        return I(f,x)
    draw(If,10**3)

def In(f,x0,n=100):
    abcisas=np.linspace(min(x0),max(x0),n)
    In=[]
    for i in range(len(abcisas)):
        In.append(I(f,abcisas[i]))
    return abcisas,In

def En(f,x0,n=100):
    abcisas=np.linspace(min(x0),max(x0),n)
    Error=[]
    for i in range(len(abcisas)):
        Error.append(f(abcisas[i])-I(f,abcisas[i]))
    return abcisas,Error

def equis(N):
    x0=[]
    for j in range(N+1):
        x0.append(-1+(2/N)*j)
    return x0
def f1(x):
    return m.sin(m.pi*x)
def f2(x):
    return 1/(1+25*x**2)
```
