```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
x0 = np.linspace(-1,1,5)
q = len(x0)-1
def li(i,x):
    l=1
    for k in range(len(x0)):
       if i!=k:
        l*= (x-x0[k])/(x0[i]-x0[k])
    return l

def I(func,x):
    I=0
    for i in range(len(x0)):
       I+= func(x0[i]) * li(i,x)
    return I

def PI(x):
    p = 1
    for i in range(len(x0)):
        p*=(x-x0[i])
    return p

def Leb(x):
    L = 0
    for i in range(len(x0)):
        L+= abs(li(i,x))
    return L

def prima(f,x,eps=10**(-6)):
    return (f(x+eps)-f(x))/eps

def primaLeb(x):
    return prima(Leb,x)

def primaPI(x):
    return prima(PI,x)

def If(x):
    return I(f,x)
x=sp.Symbol('x')
x1=sp.Symbol('x1')
x2=sp.Symbol('x2')
x3=sp.Symbol('x3')
x4=sp.Symbol('x4')


#Apartado 1 (en papel), derivadas anal√≠nicas
l0 = (x-x1)*(x-x2)*(x-x3)*(x-x4) #es el numerador solo, el denominador se convierte en deltas de x
# print(sp.diff(l0,x))

#Apartado 2 (en papel), expresiones de error

#Apartado 3, representar:
    #a) pi'(x)
    #b) pi''(x)
    #c) Leb'(x)
    #d) Leb''(x)
xd = np.linspace(-1,1,10**5)
pid = PI(xd)
pi1d = primaPI(xd)
pi2d = prima(primaPI,xd)
ld = Leb(xd)
l1d = primaLeb(xd)
l2d = prima(primaLeb,xd)

plt.title(r"$\pi_{}'$ y $\pi_{}''$".format(q+1,q+1))
plt.plot(xd,pid,'--',label=r"$\pi_{}$".format(q+1))
plt.plot(xd,pi1d,label=r"$\pi_{}'$".format(q+1))
plt.plot(xd,pi2d,label=r"$\pi_{}''$".format(q+1))
plt.legend()
plt.show()

plt.title(r"$\lambda_{}$ y $\lambda_{}''$".format(q,q))
plt.plot(xd,ld,'--',label=r"$\lambda_{}$".format(q))
plt.plot(xd,l1d,label=r"$\lambda_{}'$".format(q))
plt.plot(xd,l2d,label=r"$\lambda_{}''$".format(q))
plt.legend()
plt.show()

#Apartado 4
e0 = 10**-3
def e(x):
    return e0
def f(x):
    return np.cos(x) + e(x)

```
