```python
import matplotlib.pyplot as plt
import math as m
import numpy as np
def dibujar(f,a=-1,b=1,n=10**3):
    x=np.linspace(a,b,n)
    y=np.linspace(a,b,n)
    for i in range(n):
        y[i]=f(x[i])
    plt.plot(x,y)

def xy(f,a=-1,b=1,n=10**3):
    x=np.linspace(a,b,n)
    y=np.linspace(a,b,n)
    for i in range(n):
        y[i]=f(x[i])
    return x,y


def dibujarDsh(f,a=-1,b=1,n=10**3):
    x=np.linspace(a,b,n)
    y=np.linspace(a,b,n)
    for i in range(n):
        y[i]=f(x[i])
    plt.plot(x,y,'--')


def x(N):
    x0=[]
    for j in range(N+1):
        x0.append(-1+2*j/N)
    return x0

def y(f,x0):
    y0=[]
    for j in range(N+1):
        y0.append(f(x0[j]))
    return y0

def l(i,x):
    li=1
    for j in range(len(x0)):
        if j!=i:
            li*=(x-x0[j])/(x0[i]-x0[j])
    return li
def I(f,x):
    In=0
    for i in range(len(x0)):
        In+=l(i,x)*f(x0[i])
    return In

def f1(x):
    return m.sin(m.pi*x)

def f2(x):
    return 1/(1+25*x**2)

def If1(x):
    return I(f1,x)
def If2(x):
    return I(f2,x)

#Dibujamos f1 y su interpolante para varios N
fig1=plt.figure()
plt.title('Interpolante de $f_{1}(x)$ para varios valores de N')
dibujarDsh(f1)
leyenda=[]
for i in range(3):
    leyenda.append('')
    N=2+i*2
    x0=x(N)
    y0=y(f1,x0)
    Ix,Iy = xy(If1)
    leyenda[i],=plt.plot(Ix,Iy,label='N = '+str(N))
fig1.legend(handles=leyenda);
plt.show()

fig2=plt.figure()
plt.title('Interpolante de $f_{2}(x)$ para varios valores de N')
leyenda=[]
dibujarDsh(f2)
for i in range(3):
    leyenda.append('')
    N=3+i*4
    x0=x(N)
    y0=y(f2,x0)
    Ix,Iy = xy(If2)
    leyenda[i],=plt.plot(Ix,Iy,label='N = '+str(N))
fig2.legend(handles=leyenda);
plt.show()
```

```python
#Error del interpolante

def E(f,x):
    return f(x)-I(f,x)

def Ef1(x):
    return E(f1,x)
def Ef2(x):
    return E(f2,x)

fig1=plt.figure()
plt.title('Error del interpolante de $f_{1}(x)$ en funcin de N')
leyenda=[]
for i in range(3):
    N=3+2*i
    x0=x(N)
    equis,ygriega = xy(Ef1)
    leyenda.append('')
    leyenda[i],=plt.plot(equis,ygriega,label='N = '+str(N))
plt.legend(handles=leyenda)
plt.show()

fig2=plt.figure()
plt.title('Error del interpolante de $f_{2}(x)$ en funcin de N')
leyenda=[]
for i in range(4):
    N=4+2*i
    x0=x(N)
    equis,ygriega = xy(Ef2)
    leyenda.append('')
    leyenda[i],=plt.plot(equis,ygriega,label='N = '+str(N))
plt.legend(handles=leyenda)
plt.show()
```

```python
#Funcin pi_{n+1}

def PI(x):
    p=1
    for i in range(len(x0)):
       p*=(x-x0[i])
    return p

def Leb(x):
    leb=0
    for j in range(N+1):
        leb+=abs(l(j,x))
    return leb

figPI=plt.figure()
leyenda=[]
plt.title(r"Funcin $\pi_{n+1}(x)$ para varios valores de N");
for i in range(4):
    leyenda.append('')
    N=4+4*i
    x0=x(N)
    equis,ygriega = xy(PI)
    leyenda[i],=plt.plot(equis,ygriega,label='N = '+str(N))

plt.legend(handles=leyenda)
plt.show()

#Funcion de Lebesque lamda(x)=sumatorio(abs(lj))
figLeb=plt.figure()
leyenda=[]
plt.title(r"Funcin de Lebesque $\lambda(x)$ para varios valores de N")

for i in range(4):
    leyenda.append('')
    N=4+2*i
    x0=x(N)
    equis,ygriega = xy(Leb)
    leyenda[i],=plt.plot(equis,ygriega,label='N = '+str(N))
plt.legend(handles=leyenda)
plt.show()
    
```
